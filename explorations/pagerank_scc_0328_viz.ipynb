```python
# df = profile_results.csv
import pandas as pd
df = pd.read_csv('profile_results.csv')
df
```

```python
df_scc = df[(df['method'] == 'strongly connected components') & (df['tool'].str.contains('20230328-pagerank-scc-directed-only'))]
df_scc
```

```python
df_pr = df[(df['method'] == 'page rank') & (df['tool'].str.contains('20230328-pagerank-scc-directed-only'))]
df_pr
```

```python
# %pip install jupyter-utils-tddschn
# from jupyter_utils_tddschn import notebook_importer
# from graph_info_viz import get_pretty_graph_name
from utils_other import get_pretty_graph_name, get_pretty_tool_name
def prettify_df(df: pd.DataFrame) -> pd.DataFrame:
    # replace dataset (graph) and tool cols with pretty cols
    df['dataset'] = df['dataset'].apply(get_pretty_graph_name)
    df['tool'] = df['tool'].apply(get_pretty_tool_name)
    # drop cols with only NaNs
    df = df.dropna(axis=1, how='all')

    return df
df_pr = prettify_df(df_pr)
df_scc = prettify_df(df_scc)
df_pr
```

```python
from config import graph_info_json_path
import json
gi = json.loads(graph_info_json_path.read_text())
```

```python
def compare_igraph_and_easygraph(df: pd.DataFrame) -> None:
    # for each dataset, is there instances where easygraph's average_time is lower than igraph's?
    # if igraph's is lower, by what %? 
    # use this formula:
    # (slower_time - faster_time) / slower_time
    # i.e. 1 - faster_time / slower_time
    # convert to percentage, and keep 2 decimal points
    # generate code to answer this question
    for dataset in df['dataset'].unique():
        try:
            is_directed = gi[dataset]['is_directed']
            df_dataset = df[df['dataset'] == dataset]
            if df_dataset[df_dataset['tool'] == 'easygraph']['average_time'].values[0] < df_dataset[df_dataset['tool'] == 'igraph']['average_time'].values[0]:
                # print(f"{dataset} is faster on easygraph")
                print(f"easygraph is {round(100 * (1 - df_dataset[df_dataset['tool'] == 'easygraph']['average_time'].values[0] / df_dataset[df_dataset['tool'] == 'igraph']['average_time'].values[0]), 2)}% faster than igraph on {dataset}{' (Directed)' if is_directed else ''}")
            else:
                # print(f"{dataset} is faster on igraph")
                print(f"igraph is {round(100 * (1 - df_dataset[df_dataset['tool'] == 'igraph']['average_time'].values[0] / df_dataset[df_dataset['tool'] == 'easygraph']['average_time'].values[0]), 2)}% faster than easygraph on {dataset}{' (Directed)' if is_directed else ''}")
        except:
            print(f'skipping {dataset}')
compare_igraph_and_easygraph(df_pr)
```
