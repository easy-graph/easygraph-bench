```python
# df = profile_results.csv
import pandas as pd
df = pd.read_csv('profile_results.csv')
df
```

```python
# filter df so that it only contains the rows where method = 'k-core' and tool contains '0327', save as df_kcore
df_kcore = df[(df['method'] == 'k-core') & (df['tool'].str.contains('0327'))]
df_kcore
```

```python
# %pip install jupyter-utils-tddschn
# from jupyter_utils_tddschn import notebook_importer
# from graph_info_viz import get_pretty_graph_name
from utils_other import get_pretty_graph_name, get_pretty_tool_name
# replace dataset (graph) and tool cols with pretty cols
df_kcore['dataset'] = df_kcore['dataset'].apply(get_pretty_graph_name)
df_kcore['tool'] = df_kcore['tool'].apply(get_pretty_tool_name)
# drop cols with only NaNs
df_kcore = df_kcore.dropna(axis=1, how='all')

df_kcore
```

```python
import pandas as pd

# Custom function to apply background color to rows
def alternate_row_colors(row):
    if row.name % 4 < 2:
        return ['background-color: blue' for _ in row]
    else:
        return ['background-color: red' for _ in row]

# Apply the custom function using the style.apply method
styled_df_kcore = df_kcore.reset_index().style.apply(alternate_row_colors, axis=1)
styled_df_kcore
```
