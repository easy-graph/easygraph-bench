```python
import pandas as pd

# read the DataFrame from the HDF5 file
with pd.HDFStore('graph_info.h5', mode='r') as store:
    gi = store.get('graph_info')
```

```python
gi_directed = gi[gi['is_directed']]
gi_directed
```

```python
gi_er = gi[gi.index.str.startswith('er')]
gi_er
```

```python
gi_others = gi[~gi.index.str.startswith('er')]
gi_others

gi_others_lcc = gi_others[gi_others.index.str.endswith('lcc')]
gi_others_whole = gi_others[~gi_others.index.str.endswith('lcc')]
gi_others_whole_undirected = gi_others_whole[~gi_others_whole['is_directed']]
gi_others_whole_without_cheminformatics = gi_others_whole[~gi_others_whole.index.str.contains('cheminformatics')]
gi_others_whole_without_cheminformatics_or_sampled = gi_others_whole_without_cheminformatics[~gi_others_whole_without_cheminformatics.index.str.contains('sampled')]
gi_others_whole_without_cheminformatics_or_sampled
```

```python
def plot_graphs_as_bubbles(gi: pd.DataFrame, dpi: int = 300):
    # visualize values of nodes as bubbles, don't use plotly
    import matplotlib.pyplot as plt
    plt.gcf().set_dpi(dpi)
    # if is_directed is True, use red
    # if is_directed is False, use blue
    # if is_directed is None, use gray
    colors = ['red' if is_directed else 'blue' if is_directed is False else 'gray' for is_directed in gi['is_directed']]
    plt.scatter(gi['nodes'], gi['density'], s=gi['nodes']/1000, alpha=0.5, c=colors)
    # plt.scatter(gi['nodes'], gi['density'], s=gi['nodes']/1000, alpha=0.5)
    plt.xlabel('nodes')
    plt.ylabel('density')
    # use log scale for both axes
    plt.xscale('log')
    plt.yscale('log')
    # annotate graph names (i.e. the index of gi)
    # use thin line to point the annotation texts to the bubbles
    # make the annotation text fonts small
    for i, graph_name in enumerate(gi.index):
        plt.annotate(graph_name, (gi['nodes'][i], gi['density'][i]), xytext=(gi['nodes'][i]*1.1, gi['density'][i]*1.1), arrowprops=dict(arrowstyle='->', color='gray', alpha=0.5), fontsize=6)


    plt.show()
```

```python
plot_graphs_as_bubbles(gi_others_whole_without_cheminformatics_or_sampled)
```

```python
plot_graphs_as_bubbles(gi_er)
```

```python
plot_graphs_as_bubbles(gi_others_whole)
```
