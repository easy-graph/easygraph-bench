```python
# df = profile_results.csv
import pandas as pd
df = pd.read_csv('profile_results.csv')
df
```

```python
from config import graph_info_json_path
gi = pd.read_json(graph_info_json_path)
gi

# swap the col and row of gi
gi = gi.T
gi
```

```python
# filter gi so that the index of the rows must contain any of the values in the 'dataset' col of df
gi = gi[gi.index.str.contains('|'.join(df['dataset']))]
gi
```

```python
# rename values in the 'tool' col of df, only keep the part before the first '_'
df['tool'] = df['tool'].str.split('_').str[0]
df
```

```python
# remove cols in df that only contains NaN
df = df.dropna(axis=1, how='all')
df
```

```python
# also drop the 'iteration_count' col
df = df.drop('iteration_count', axis=1)
df
```

```python
# for each value of dataset in df, draw a bar plot, hue=tool, y=average_time, x=method
import seaborn as sns
import matplotlib.pyplot as plt
for dataset in df['dataset'].unique():
    sns.barplot(x='method', y='average_time', hue='tool', data=df[df['dataset'] == dataset])
    plt.title(dataset)
    plt.show()
```

```python
import numpy as np

# Function to add percentage annotations on top of each bar
def add_percentage_annotations(ax, igraph_data, easygraph_data):
    for p, igraph_val, easygraph_val in zip(ax.patches, igraph_data, easygraph_data):
        height = p.get_height()
        if height > 0:
            ax.annotate('{:.1%}'.format(igraph_val / easygraph_val),
                        (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='baseline',
                        size=12, xytext=(0, 5),
                        textcoords='offset points')

# for each value of 'method' in df, draw a bar plot, hue=tool, y=average_time, x=dataset
for method in df['method'].unique():
    # Create the bar plot
    ax = sns.barplot(x='dataset', y='average_time', hue='tool', data=df[df['method'] == method])

    # Get the average_time values for igraph and easygraph for each dataset
    igraph_data = df[(df['method'] == method) & (df['tool'] == 'igraph')]['average_time'].values
    easygraph_data = df[(df['method'] == method) & (df['tool'] == 'easygraph')]['average_time'].values

    # Call the function to add the percentage annotations
    add_percentage_annotations(ax, igraph_data, easygraph_data)

    # Set the title and show the plot
    plt.title(method)
    plt.show()
```
